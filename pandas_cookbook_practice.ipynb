```python
"""
use the college dataset to calculate the mean and standard deviation of the
undergraduate student population per state. We then use this information to find the
maximum number of standard deviations from the mean that any single population value
is per state.
"""

data.groupby(['STABBR']).agg({'UGDS': ['mean', 'std']}).round(2)
```

```python
def max_deviation(s):
    std_score = (s - s.mean()) / s.std()
    
    return std_score.abs().max()
```

```python
data.groupby(['STABBR'])['UGDS'].agg(max_deviation).round(2)
```

```python
max_deviation.__name__ = 'Max Deviation'

data.groupby(['STABBR', 'RELAFFIL'])['UGDS', 'SATVRMID', 'SATMTMID'].agg([max_deviation, 'mean', 'std']).head()
```

```python
def pct_btw_1_and_3k(s):
    return s.between(1000, 3000).mean()
```

```python
data.groupby(['STABBR', 'RELAFFIL'])['UGDS'].agg(pct_btw_1_and_3k).head(10)
```

```python
#Altering the funtion

def pct_between(s, low, high):
    return s.between(low, high).mean()
```

```python
data.groupby(['STABBR', 'RELAFFIL'])['UGDS'].agg(pct_between, 1000, 10000).head(10)
```

```python
data.groupby(['STABBR', 'RELAFFIL'])['UGDS'].agg([pct_between,1000,10000,'mean']).head()
```

```python
#Using *args, **kwargs                                #Funtion Closure

def make_agg_func(func, name, *args, **kwargs):
    def wrapper(x):
        return func(x, *args, **kwargs)
    wrapper.__name__ = name
    return wrapper
```

```python
my_agg1 = make_agg_func(pct_between, 'pct_btw_1_3k', 1000, 3000)
my_agg2 = make_agg_func(pct_between, 'pct_btw_1_10k', 1000, 10000)
```

```python
data.groupby(['STABBR', 'RELAFFIL'])['UGDS'].agg(['mean',my_agg1, my_agg2]).round(2).head()
```

```python
weight_loss = pd.read_csv('weight_loss.csv')
```

```python
weight_loss
```

```python
weight_loss.query('Month == "Jan"')
```

```python
def find_perc_loss(s):
    return (s - s.iloc[0])/ s.iloc[0]
```

```python
bob_jan = weight_loss.query('Name == "Bob" and Month =="Jan"')
find_perc_loss(bob_jan['Weight'])
```

```python
pct_loss = weight_loss.groupby(['Name', 'Month'])['Weight'].transform(find_perc_loss)
```

```python
pct_loss.head()
```

```python
weight_loss['Perc Weight Loss'] = pct_loss.round(3)
weight_loss.query('Name=="Bob" and Month in ["Jan", "Feb"]')
```

```python
week4 = weight_loss.query('Week == "Week 4"')
```

```python
week4
```

```python
winner = week4.pivot(index='Month', columns='Name',values='Perc Weight Loss')
```
